```python
skills = [
    'Python',
    'SQL',
    'Excel',
    'Power BI',
    'Tableau',
    'Machine Learning',
    'Artificial Intelligence',
    'AI',
    'Git',
    'JavaScript',
    'Java',
    'C++',
    'Communication',
    'Problem Solving',
    'Data Analysis',
    'Reporting',
    'Automation'
]
```

```python
for skill in skills:
    df[skill] = df['Full Job Description'].str.contains(
        skill,
        case=False,
        na=False
    ).astype(int)

print(df.head())
```

```python
skill_counts = df[skills].sum().sort_values(ascending=False)

print(skill_counts)
```

```python
print(skill_counts)
```

```python
import re

for skill in skills:
    if skill == "C++":
        pattern = r"\bC\+\+\b"
    elif skill == "AI":
        pattern = r"\bAI\b"
    else:
        pattern = rf"\b{re.escape(skill)}\b"

    df[skill] = df['Full Job Description'].str.contains(
        pattern,
        case=False,
        na=False,
        regex=True
    ).astype(int)
```

```python
skill_counts = df[skills].sum().sort_values(ascending=False)

print(skill_counts)
```

```python
print(skill_counts)
```

```python
import matplotlib.pyplot as plt

skill_counts.plot(kind='bar')

plt.title('Top Skills in Entry-Level South African Tech Roles')
plt.xlabel('Skills')
plt.ylabel('Number of Job Posts')

plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

# Recommendations for AGN Internship Programme

## Findings

This project analysed 10 entry-level South African technology job postings collected from public job platforms such as LinkedIn and Indeed. The dataset was cleaned using Python and pandas by checking for missing values, removing duplicate job postings, and standardising locations, work types, and role categories. Skills were then extracted from job descriptions to identify common requirements in the South African entry-level technology job market.

The analysis indicated that AI-related skills appeared in 4 out of the 10 job postings analysed. Python, SQL, Automation, and Reporting were each mentioned in 2 job postings. Other skills, including Machine Learning, Tableau, Power BI, JavaScript, and Data Analysis, appeared once. These findings suggest that employers increasingly value a combination of programming, analytical, reporting, and emerging AI-related competencies.

Although this analysis was based on a relatively small sample of job postings, it provides useful insights into emerging skill requirements within the South African technology sector.

## Recommendations

Based on the findings, AGN should prioritise the following areas in its internship programme:

1. **Artificial Intelligence and Automation**

   * Introduce interns to AI concepts, automation tools, and practical AI applications while strengthening their understanding of how AI is used in modern workplaces.

2. **Python Programming**

   * Focus on Python for data analysis, automation, scripting, and problem-solving, as it remains one of the most versatile and in-demand programming languages.

3. **SQL and Database Skills**

   * Provide training in database fundamentals, data management, and SQL querying techniques to help interns work effectively with structured data.

4. **Data Visualisation and Reporting**

   * Equip interns with skills in Power BI, Tableau, and dashboard development to enable them to communicate insights effectively.

5. **Hands-on Projects and Practical Experience**

   * Incorporate real-world projects and case studies that simulate workplace scenarios, allowing interns to apply technical knowledge while developing analytical and critical thinking skills.

## Limitations

This analysis was based on a sample of 10 job postings collected over a limited period from selected job platforms. As a result, the findings may not fully represent the entire South African entry-level technology job market. Future studies could analyse a larger dataset collected over a longer period to provide more comprehensive insights into industry skill demands.

## Conclusion

The findings suggest that South African entry-level technology roles increasingly require skills related to AI, programming, data analysis, reporting, and data visualisation. By aligning its internship programme with these emerging industry requirements, AGN can better equip interns with practical and relevant skills that improve their employability and readiness for the evolving technology sector.
